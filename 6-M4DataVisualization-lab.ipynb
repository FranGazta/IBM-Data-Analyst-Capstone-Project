## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
QUERY = """
SELECT convertedcomp
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(10, 5))
sns.distplot(df.dropna(), hist_kws={'edgecolor':'black','alpha':0.5}, kde=False)
plt.xlabel('Salary')
plt.ylabel('Count')
plt.title('Salary dsitribution', size=16)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(8, 4))
sns.boxplot(df.dropna(), whiskerprops=dict(linestyle='--'), fliersize=4)
plt.xlabel('Age')
plt.title('Age dsitribution', size=14)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 5))
sns.scatterplot(df['Age'],df['WorkWeekHrs'], edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Hours worked/week')
plt.title('Age vs. hours worked weekly', size=16)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 5))
sns.scatterplot(data = df, y = 'WorkWeekHrs', x = 'CodeRevHrs', size = 'Age', alpha = 0.6, sizes=(1, 600))
plt.title('Relation between Age worked week hours and', size=16)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *,COUNT(*) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count desc
"""
df = pd.read_sql_query(QUERY,conn)

data=df['count'][0:5]
labels=df['DatabaseDesireNextYear'][0:5]
plt.figure(figsize=(10, 5))
plt.pie(data, labels=labels, autopct='%1.1f%%', wedgeprops={'edgecolor': 'gray'}, colors = ["#87CEEB", "#90EE90", "#F08080", "#FFA07A", "#FFFFE0"])
plt.title('The 5 most wished-learn databases', size=16)
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 and 35
"""
df = pd.read_sql_query(QUERY,conn)
wwh = df['WorkWeekHrs'].quantile(0.5)
crh = df['CodeRevHrs'].quantile(0.5)

fig, ax = plt.subplots()

ax.bar('Age 30 - 35', wwh, label='WorkWeekHrs')
ax.bar('Age 30 - 35', crh, bottom=wwh, label='CodeRevHrs')
ax.legend(loc='lower right')
plt.ylabel('Hours')
plt.title('Accumulated hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 and 60
"""
df = pd.read_sql_query(QUERY,conn)

medians = df.groupby('Age')['ConvertedComp'].median()

sns.lineplot(data=medians)
plt.xlabel('Age')
plt.ylabel('Median compensations')
plt.title('Median compensation by age', size=15)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
nuevo=df['MainBranch'].value_counts()
nuevo=pd.DataFrame(nuevo)
plt.barh([textwrap.fill(cat, width=15) for cat in nuevo.index], nuevo['MainBranch'], color=["#87CEEB", "#90EE90"])
plt.xlabel('People')
plt.ylabel('Categories')
plt.title('Primary job or part time job?')
plt.show()
```
